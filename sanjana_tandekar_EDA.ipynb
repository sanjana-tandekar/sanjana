```python
# Convert 'TransactionDate' to datetime format
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'], errors='coerce')

# Create 'MonthYear' column in the format "YYYY-MM"
transactions['MonthYear'] = transactions['TransactionDate'].dt.to_period('M').astype(str)

transactions['TotalValue'] = pd.to_numeric(transactions['TotalValue'], errors='coerce')
transaction_trends = transactions.groupby('MonthYear')['TotalValue'].sum()
plt.figure(figsize=(10, 6))
transaction_trends.plot(kind='line', marker='o', color='blue')
plt.title("Transaction Value Trends Over Time")
plt.xlabel("Month-Year")
plt.ylabel("Total Value")
plt.grid(True)
plt.show()
```

```python
combined = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")
```

```python
top_customers = combined.groupby("CustomerID")["TotalValue"].sum().sort_values(ascending=False).head(10)
print("Top 10 Customers by Revenue:")
print(top_customers)
category_revenue = combined.groupby("Category")["TotalValue"].sum().sort_values(ascending=False)

```

```python
plt.figure(figsize=(8, 6))
category_revenue.plot(kind='bar', color='skyblue')
plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python

# Convert dates to datetime format
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

# Merge datasets for combined analysis
merged_data = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID', suffixes=('_trans', '_prod'))

# ===============================
# 1. Advanced Analysis of Transaction Data
# ===============================

# Average transaction value
avg_transaction_value = transactions['TotalValue'].mean()
print("\nAverage Transaction Value: ${:.2f}".format(avg_transaction_value))

# Most loyal customers (based on transaction count)
loyal_customers = transactions['CustomerID'].value_counts().head(10)
print("\nTop 10 Loyal Customers (by transaction count):")
print(loyal_customers)

plt.figure(figsize=(10, 6))
loyal_customers.plot(kind='bar', color='green')
plt.title('Top 10 Loyal Customers')
plt.xlabel('Customer ID')
plt.ylabel('Transaction Count')
plt.show()




```

```python

# ===============================
# 2. Customer-Product Analysis
# ===============================

# Popular product categories by region
popular_categories_by_region = merged_data.groupby(['Region', 'Category'])['Quantity'].sum().unstack()
print("\nPopular Product Categories by Region:")
print(popular_categories_by_region)

plt.figure(figsize=(12, 6))
sns.heatmap(popular_categories_by_region, cmap='Blues', annot=True, fmt=".0f")
plt.title('Popular Product Categories by Region')
plt.xlabel('Category')

plt.ylabel('Region')
plt.show()
```

```python
# High-value customers (customers contributing the most revenue)
high_value_customers = merged_data.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 High-Value Customers (by revenue):")
print(high_value_customers)

plt.figure(figsize=(10, 6))
high_value_customers.plot(kind='bar', color='orange')

plt.title('Top 10 High-Value Customers')
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# ===============================
# 3. Correlation Analysis
# ===============================

# Correlation between product price and quantity sold
correlation_price_quantity = merged_data[['Price_trans','Quantity']].corr()
print("\nCorrelation between Product Price and Quantity Sold:")
print(correlation_price_quantity)

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_price_quantity, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Product Price and Quantity Sold')
plt.show()

```

```python
# Pairplot of key transaction features
sns.pairplot(merged_data[['Price_trans', 'Quantity', 'TotalValue']], diag_kind='kde', corner=True)
plt.suptitle('Pairplot of Transaction Features', y=1.02)
plt.show()
```
